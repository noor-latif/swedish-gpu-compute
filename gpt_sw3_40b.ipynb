```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-generation", model="AI-Sweden-Models/gpt-sw3-40b")
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("AI-Sweden-Models/gpt-sw3-40b")
model = AutoModelForCausalLM.from_pretrained("AI-Sweden-Models/gpt-sw3-40b")
```

### Test the model with a sample input

```python
sample_input = "Hello, I am a large language model. Please tell me a story about a dragon."
# Using the pipeline for text generation
output = pipe(sample_input)
print(output[0]['generated_text'])
```
